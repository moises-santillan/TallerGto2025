```python
import library as lb
import matplotlib.pyplot as plt
```

```python
# Example 1: Solving the exponential decay model

t_end = 10
initial_condition = [1]  # Starting with y0 = 1
params = (0.5,)  # Example parameters a and b

t, y = lb.solve_differential_equation(lb.exponential_decay_model, t_end, initial_condition, params)

```

```python
plt.plot(t, y.T)
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.title('Solution of the Differential Equation')
plt.grid()
plt.show()
```
