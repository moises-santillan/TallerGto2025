```python
import library as lb
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_data=[0,20,40,60,80,100,120]
glucose_data=[1,1.525581395,1.776744186,1.469767442,1.446511628,1.446511628,1.148837209]
```

```python
absortion = 0.13
glucose_rate = .00
insulin_rate = 0.012
delay = 15

t=np.linspace(1, 120, 100)
glucose = lb.glucose_function(t, absortion, glucose_rate, insulin_rate, delay)
plt.plot(t_data, glucose_data, 'o')
plt.plot(t, glucose)
```

```python
def error(par):
    return (np.sum(lb.glucose_function(t_data, *par) - glucose_data))**2
```

```python
error((absortion, glucose_rate, insulin_rate, delay))
```

```python
from scipy.optimize import differential_evolution
```

```python
bounds = [(0, .2),(0, .05),(0, .05),(0, 20)]
result = differential_evolution(error, bounds)
popt=result.x
```
