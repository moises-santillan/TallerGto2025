# Exponential Decay

Let us analyze the exponential decay model
$$
\frac{dy}{dt}=-\mu y.
$$
First, we plot the right-hand side of the differential equation as a function of $y$ for different values of the decay rate $\mu$. Observe that, in all cases, the curves intersect axis $x$ at the origin with a negative slope, indicating that $x^*=0$ is the system unique steady state, and that it is positive for all values of $\mu$.

We observe as well that the slope gets steeper as $\mu$ increases. Therefore, the system should return sooner for higher values of $\mu$.

```python
t = 0
y = np.linspace(0,4,100)
decay_rate = [0.1, 0.2, 0.5, 1]
for dr in decay_rate:
    params = (dr,)
    ydot = lb.exponential_decay_model(t, y, *params)
    plt.plot(y, ydot, label=fr"$\mu = {dr}$")
plt.axhline(y=0, color="gray")
plt.grid()
plt.legend()
plt.xlabel(r'$y$')
plt.ylabel(r'$dy/dt$')
```

We now solve numerically the differential equations with different values of the decay rate and different initial conditions, and plot the solutions. The results are consistent with the stability analysis.

```python
# Example 1: Solving the exponential decay model

t_end = 20
initial_condition = [1]  # Starting with y0 = 1
decay_rate = [0.1, 0.2, 0.5, 1]

for dr in decay_rate:
    params = (dr,)  # Example parameters a and b
    t, y = lb.solve_differential_equation(lb.exponential_decay_model, t_end, initial_condition, params)
    plt.plot(t, y.T, label=fr"$\mu = {dr}$")
plt.grid()
plt.legend()
plt.xlabel(r'$t$')
plt.ylabel(r'$y(t)$')
```

```python
# Example 1: Solving the exponential decay model

t_end = 20
initial_condition = [2]  # Starting with y0 = 2
decay_rate = [0.1, 0.2, 0.5, 1]

for dr in decay_rate:
    params = (dr,)  # Example parameters a and b
    t, y = lb.solve_differential_equation(lb.exponential_decay_model, t_end, initial_condition, params)
    plt.plot(t, y.T, label=fr"$\mu = {dr}$")
plt.grid()
plt.legend()
plt.xlabel(r'$t$')
plt.ylabel(r'$y(t)$')
```

# Mathematical model for gene expression with positive feedback regulation.

Consider this model:
$$
\frac{dy}{dt} = \alpha + \beta \frac{y^4}{y^4 + 1} + y,
$$
where $\alpha$ represents the basal gene expression rate and $\alpha+\beta$ is the maximum gene expression rate.

Below, the right-hand side of differential equation is plotted vs. $y$ and the corresponding numerical solution of the differential equation is plotted vs. $t$ for sample parameter values and a specific initial condition. 

```python
t = 0
y = np.linspace(0,4,100)
basal_expression = 0.3
maximal_expression = 2
params = (basal_expression, maximal_expression)
ydot = lb.gene_expression_model(t, y, *params)
plt.plot(y, ydot)
plt.axhline(y=0,color='gray')
plt.grid()
plt.xlabel(r'$y$')
plt.ylabel(r'$dy/dt$')
```

```python
# Example 1: Solving the logistic growth model

t_end = 50
initial_condition = [.5]  # Starting with y0 = .5
basal_expression = 0.3
maximal_expression = 2
params = (basal_expression, maximal_expression)

t, y = lb.solve_differential_equation(lb.logistic_growth_model, t_end, initial_condition, params)

plt.plot(t, y.T)
plt.xlabel(r'$t$')
plt.ylabel(r'$y(t)$')
plt.grid()
```
